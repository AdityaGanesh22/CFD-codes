# Domain Vertex method for Structured grid problems

With the help of this method, we can get the solution for a complex geometry problem that cannot be solved in regular rectangular coordinates. A rectangular grid, known as the computational domain is first constructed and then the physical domain is mapped onto it. This makes it possible to get accurate solutions at the points of interest. In this notebook, we will attempt to solve the Laplace equation for a structured grid.

\begin{equation} \label{Laplace}
    \nabla^2 T = 0
\end{equation}

### We first import the libraries that will be useful for us
Numpy = Used for creating arrays, matrices, solving Matrix equations, etc. \
Matplotlib = Used to visualize graphs, contour plots, etc. \
Sympy = Used to symbolically store equations as functions

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy.abc import xi, eta
%matplotlib inline
```

### Step 1: Declaring Inputs
To begin the Domain Vertex Method, we first have to declare the inputs to describe the problem. Here, we describe the number of grids required in the $\xi$ and $\eta$ direction. As we also want to solve the Heat transfer equation, we define the boundary conditions for the problem. As the discretization is done using quadratic Lagrange interpolation, we define the points on the boundary as shown above.

```python
p = 40                  # Divisions in Xi direction
q = 40                  # Divisions in Eta direction
T_left = 1000           # Temperature on Left Boundary, units
T_right = 100           # Temperature on Right Boundary, units
T_top = 500             # Temperature on Top Boundary, units
T_bottom = 100          # Temperature on Bottom Boundary, units

# points ->          1        2        3        4        5         6        7        8
points = np.array([[2, 1], [11, 5], [21, 2], [19, 9], [22, 20], [12, 23], [4, 20], [1, 8]])
# points = np.array([[0, 0], [10, 0], [20, 0], [20, 5], [20, 10], [10, 7.5], [0, 5], [0, 2.5]])
# points = np.array([[1.464, 1.464], [5, 0], [8.536, 1.464], [10, 5], [8.536, 8.536], [5, 10], [1.464, 8.536], [0, 5]])
# points = np.array([[5, 10], [10, 7.8], [20, 7.1], [40, 6.5], [60, 6], [40, 10], [20, 15], [8, 15]])
```

Parameters that can be deduced from the inputs are as follows:-

```python
Xi = np.linspace(0, 1, p)
Eta = np.linspace(0, 1, q)
Delta_xi = Xi[1] - Xi[0]
Delta_eta = Eta[1] - Eta[0]
n = p * q
G = np.zeros([n, n])
H = np.zeros([n, 1])
```

### Step 2: Lagrange Blending Functions

The Multidimensional interpolation for DVM is done by taking the tensor product of two unidirectional interpolating functions. We use the Lagrange Polynomials as the interpolating functions, and the tensor product of both the functions in two different directions give rise to blending functions. The 8 - node Lagrange Blending functions is directly stated as below. (Ref. [1]) \
\
(You may also consider skipping steps 2 and 3 if you want to manually enter the spatial functions.)

```python
phi = sp.Array([(1 - xi) * (1 - eta) * (1 - 2 * xi - 2 * eta),  # phi_1
                4 * xi * (1 - xi) * (1 - eta),                  # phi_2
                xi * (1 - eta) * (2 * xi - 2 * eta - 1),        # phi_3
                4 * xi * eta * (1 - eta),                       # phi_4
                - xi * eta * (3 - 2 * xi - 2 * eta),            # phi_5
                4 * xi * (1 - xi) * eta,                        # phi_6
                (1 - xi) * eta * (2 * eta - 2 * xi - 1),        # phi_7
                4 * eta * (1 - xi) * (1 - eta)])                # phi_8
```

### Step 3: Calculating functions $F_1$ and $F_2$ which store the $x$ and $y$ spatial information
The blending functions, along with the points that have been declared as an input are used to formulate these multidirectional interpolating functions

```python
F1, F2 = 0, 0
for i in range(len(phi)):
    F1 += phi[i] * points[i, 0]
    F2 += phi[i] * points[i, 1]
```

### Step 4: Calculating the x and y co-ordinates at the discrete points using the above functions
Two matrices are created to store the x and y coordinates that are generated from the above functions. In case one wishes to use their own functions, one can execute this section directly from Step 1 to solve the Heat Transfer Equation.

```python
x = np.zeros([q, p])
y = np.zeros([q, p])
for i in range(q):
    for j in range(p):
        x[i, j] = F1.subs({xi:Xi[j], eta:(1 - Eta[i])})       # (2 + 2 * Xi[i]) * np.cos(np.pi/2 * Eta[j])
        y[i, j] = F2.subs({xi:Xi[j], eta:(1 - Eta[i])})       # (2 + 2 * Xi[i]) * np.sin(np.pi/2 * Eta[j])
```

### Step 5: Plotting the grid (optional)

```python
def GridPlot(x, y, plot_title, x_label, y_label):
    for i in range(len(x[1,:])):
        plt.plot(x[:, i], y[:, i], 'k')
    for j in range(len(x[:,1])):
        plt.plot(x[j, :], y[j, :], 'k')
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    return plt.show()
    
[mesh_Xi, mesh_Eta] = np.meshgrid(Xi, Eta)
GridPlot(mesh_Xi, 1 - mesh_Eta,'Computational Domain', r'$\xi \rightarrow$', r'$\eta \rightarrow$')
GridPlot(x, y,'Physical Domain', r'$x \rightarrow$', r'$y \rightarrow$')
```

### Step 6: Transformation of Physical Domain ($x-y$) coordinates into Computational Domain ($\xi - \eta$) coordinates
As Laplace Equation cannot be comprehended in the physical domain, it is transformed and solved in cartesian coordinates. (Ref. [2])

```python
J = sp.Matrix([F1, F2]).jacobian(sp.Matrix([xi, eta]))
Beta = J.cofactor_matrix()

B = sp.zeros(Beta.shape[0], Beta.shape[1])
for j in range(Beta.shape[0]):
    for k in range(Beta.shape[1]):
        count = 0
        for i in range(Beta.shape[0]):
            count += Beta[i, j] * Beta[i, k]
        B[j,k] = count

C = B/J.det()
```

### Step 7: Solution of the Laplace equation using Finite Difference Scheme in $\xi-\eta$ plane

```python
# Left and Right Boundary Conditions
for i in range(q):
    # Left
    G[p * i,p * i] = 1
    H[p * i] = T_left
    # Right
    G[p * i - 1, p * i - 1] = 1
    H[p * i - 1] = T_right

# Top and Bottom Boundary Conditions
for j in range(p):
    # Top
    if H[j, 0] != 0 and G[j, j] == 1:
        H[j] = (H[j] + 500) / 2
    else:
        G[j, j] = 1
        H[j] = T_top
    # Bottom
    if H[n - p + j, 0] != 0 and G[n - p + j, n - p + j] == 1:
        H[n - p + j] = (H[n - p + j] + 100)/2
    else:
        G[n - p + j, n - p + j] = 1
        H[n - p + j] = T_bottom

# # Internal grids
for i in range(1, q - 1):
    for j in range(1, p - 1):
        Dummy = np.zeros([q, p])
        Dummy[i, j] = (C[0, 0].subs({xi:Xi[j + 1], eta:Eta[i]}) + 2 * C[0, 0].subs({xi:Xi[j], eta:Eta[i]}) + C[0, 0].subs({xi:Xi[j - 1], eta:Eta[i]})) / (2 * Delta_xi**2) + \
                      (C[1, 1].subs({xi:Xi[j], eta:Eta[i + 1]}) + 2 * C[1, 1].subs({xi:Xi[j], eta:Eta[i]}) + C[1, 1].subs({xi:Xi[j], eta:Eta[i - 1]})) / (2 * Delta_eta**2)
        Dummy[i + 1, j] = - (C[0, 0].subs({xi:Xi[j + 1], eta:Eta[i]}) + C[0, 0].subs({xi:Xi[j], eta:Eta[i]})) / (2 * Delta_xi**2)
        Dummy[i - 1, j] = - (C[0, 0].subs({xi:Xi[j - 1], eta:Eta[i]}) + C[0, 0].subs({xi:Xi[j], eta:Eta[i]})) / (2 * Delta_xi**2)
        Dummy[i, j + 1] = - (C[1, 1].subs({xi:Xi[j], eta:Eta[i + 1]}) + C[1, 1].subs({xi:Xi[j], eta:Eta[i]})) / (2 * Delta_eta**2)
        Dummy[i, j - 1] = - (C[1, 1].subs({xi:Xi[j], eta:Eta[i - 1]}) + C[1, 1].subs({xi:Xi[j], eta:Eta[i]})) / (2 * Delta_eta**2)
        Dummy[i - 1, j - 1] = - (C[1, 0].subs({xi:Xi[j - 1], eta:Eta[i]}) + C[0, 1].subs({xi:Xi[j], eta:Eta[i - 1]})) / (4 * Delta_eta * Delta_xi)
        Dummy[i + 1, j + 1] = - (C[1, 0].subs({xi:Xi[j + 1], eta:Eta[i]}) + C[0, 1].subs({xi:Xi[j], eta:Eta[i + 1]})) / (4 * Delta_eta * Delta_xi)
        Dummy[i + 1, j - 1] =   (C[1, 0].subs({xi:Xi[j + 1], eta:Eta[i]}) + C[0, 1].subs({xi:Xi[j], eta:Eta[i - 1]})) / (4 * Delta_eta * Delta_xi)
        Dummy[i - 1, j + 1] =   (C[1, 0].subs({xi:Xi[j - 1], eta:Eta[i]}) + C[0, 1].subs({xi:Xi[j], eta:Eta[i + 1]})) / (4 * Delta_eta * Delta_xi)
        G[i * p + j, :] = Dummy.reshape([1, n])

Temp = np.linalg.solve(G, H).reshape([q, p])
```

### Step 8: Plotting the results

```python
def ContourPlot(x, y, z, plot_title, x_label, y_label):
    plt.contourf(x, y, z)
    plt.colorbar()
    plt.title(plot_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    # plt.xlim([min(np.min(x) * 0.99, np.min(x) * 1.01), max(np.max(x) * 1.01, np.max(x) * 0.99)])
    # plt.ylim([min(np.min(y) * 0.99, np.min(y) * 1.01), max(np.max(y) * 1.01, np.max(y) * 0.99)])
    return plt.show()

ContourPlot(Xi, 1 - Eta, Temp, 'Computational Domain', r'$\xi \rightarrow$', r'$\eta \rightarrow$')
ContourPlot(x, y, Temp, 'Physical Domain', r'$x \rightarrow$', r'$y \rightarrow$')
```
